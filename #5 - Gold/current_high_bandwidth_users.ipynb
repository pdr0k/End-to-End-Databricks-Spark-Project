```
%sql

CREATE OR REPLACE TABLE gold.current_high_bandwidth_users
LOCATION 'dbfs:/FileStore/gold/current_high_bandwidth_users'

SELECT
  IP,
  BandwidthUsage,
  Speed_Connection,
  Connection_Time,
  Device,
  Connection_Status

FROM silver.unified_connections
WHERE
  BandwidthUsage = 'High' AND
  Connection_Status = 'Online' AND
  -- what is real time for the business?
  Connection_Time >= (current_timestamp() - INTERVAL '20' MINUTE)
  
ORDER BY Speed_Connection DESC
```

```
%sql
select  *
from gold.current_high_bandwidth_users
```
